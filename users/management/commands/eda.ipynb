```python
import pandas as pd

# Load CSV file (update filename if different)
df = pd.read_csv("calls.csv",skiprows=2)  

# OR Load Excel file
# df = pd.read_excel("calls_data.xlsx")

# Display first few rows
df.head()
```

```python
df.columns
```

```python
# Rename columns with proper formatting
df.columns = [
    's_no', 'faculty_id', 'faculty_name', 'b_no', 'student_name',
    'phone_number_1', 'phone_number_2', 'extra_column', 'inter_group',
    'inter_marks', 'entrance_rank', 'call_1', 'call_2', 'call_3'
]

# Drop the extra column if it's not needed
df = df.drop(columns=['extra_column'], errors='ignore')

# Display cleaned column names
df.head()
```

```python
# Keep only the required columns
df = df[['faculty_id', 'faculty_name', 'b_no', 'student_name', 'phone_number_1', 'phone_number_2']]
```

```python
staff = pd.read_excel("staff.xlsx",skiprows=1)
```

```python
staff.columns
```

```python
# Rename columns with proper formatting
staff = staff.rename(columns={
    'S.NO.': 's_no',
    'EMP ID': 'faculty_id',
    'NAME OF THE FACULTY': 'faculty_name',
    'DEPARTMENT': 'department',
    'PHONE NUMBER': 'phone_number'
})

```

```python
staff.head()
```

```python
df.head()
```

```python
# Clean Staff Data
staff_df = staff.rename(columns={
    'S.NO.': 's_no',
    'EMP ID': 'faculty_id',
    'NAME OF THE FACULTY': 'faculty_name',
    'DEPARTMENT': 'department',
    'PHONE NUMBER': 'phone_number'
})

# Clean Students Data
students_df = df.rename(columns={
    'faculty_id': 'faculty_id',
    'faculty_name': 'faculty_name',
    'b_no': 'b_no',
    'student_name': 'student_name',
    'phone_number_1': 'phone_number_1',
    'phone_number_2': 'phone_number_2'
})

# Strip whitespace from all text columns
staff_df = staff_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
students_df = students_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Clean phone numbers in staff data (handle multiple numbers separated by "\n")
staff_df['phone_number'] = staff_df['phone_number'].str.replace(r'\n', ',', regex=True)

# Remove non-numeric characters from phone numbers in students data
students_df['phone_number_1'] = students_df['phone_number_1'].astype(str).str.replace(r'\D', '', regex=True)
students_df['phone_number_2'] = students_df['phone_number_2'].astype(str).str.replace(r'\D', '', regex=True)

# Convert phone numbers to numeric where possible
students_df['phone_number_1'] = pd.to_numeric(students_df['phone_number_1'], errors='coerce')
students_df['phone_number_2'] = pd.to_numeric(students_df['phone_number_2'], errors='coerce')


```

```python
students_df = students_df.merge(staff_df[['faculty_id', 'faculty_name']], on='faculty_name', how='left')
```

```python
students_df
```

```python
# Convert faculty_id to string and fill missing values
students_df['faculty_id'] = students_df['faculty_id_y'].astype('Int64').astype(str)
students_df['faculty_id'] = students_df['faculty_id'].replace('<NA>', '', regex=True)  # Handle missing values
students_df['phone_number_2'] = students_df['phone_number_2'].apply(lambda x: str(int(x)) if pd.notnull(x) else '')
# Drop the extra faculty_id_x and faculty_id_y columns
students_df = students_df.drop(columns=['faculty_id_x', 'faculty_id_y'], errors='ignore')


```
